```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Nov  1 13:17:59 2019

@author: smullally
"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ["Dec 2018","Jan 2019","Feb 2019","Mar 2019","Apr 2019", "May 2019",
          "Jun 2019", "Jul 2019","Aug 2019","Sep 2019","Oct 2019"]

mon = np.arange(0,11,1)

download = [63.4,52.6,65.47,79.6,54.7,80.81,\
            86.85, 59.89, 47.34, 65.60,116.80]

unip = [827,466,691,903,785,831,
        901,1125,1365,1338,1671]
```

```python
labs=17
fig, ax1 = plt.subplots(figsize=(7,7))

color = 'black'
ax1.set_xlabel('')
ax1.set_ylabel('Download Volume (TB)', color=color, fontsize=labs)
ax1.plot(mon, download, 'o-', color=color, label = "Download Volume")
ax1.tick_params(axis='y', labelcolor=color)
plt.ylim(30,130)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'blue'
ax2.set_ylabel('Number of Unique IP Addresses', color=color, fontsize=labs)  # we already handled the x-label with ax1
ax2.plot(mon, unip, 's--', color=color, label = "Unique IP Addresses")
ax2.plot([],[], '.-', color='black', label = "Download Volume")
ax2.tick_params(axis='y', labelcolor=color)

loc, label = plt.xticks()

plt.xticks(mon, months)
#Because we are in a subplot, you can't use xticks for rotation.

ax1.tick_params(axis='x',rotation=90, labelsize=labs)
ax1.tick_params(axis='y', labelsize=labs)
ax2.tick_params(axis='both',labelsize=labs)
plt.margins(0.2)
plt.legend(loc="upper left", fontsize=14)



fig.tight_layout()
```
